## Diving the data into Input and output variables

```python
X=df.drop(["Reached.on.Time_Y.N",'ID'],axis=1)
y=df['Reached.on.Time_Y.N']
```

```python
X.head()
```

```python
## Using one hot encoder on categorical input columns
```

```python
import category_encoders as ce

ohe=ce.OneHotEncoder(cols=['Warehouse_block','Mode_of_Shipment','Product_importance','Gender'])
X=ohe.fit_transform(X)
```

```python
X
```

# Spliting data into train test split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=3)
```

## Appling Standard scaling  

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

## knn classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)
accuracy_score(y_pred_knn,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

knn=KNeighborsClassifier()

knc_params={'n_neighbors':list(range(1,50))}

grid_search=GridSearchCV(knn,knc_params,cv=10)

grid_search.fit(X_train,y_train)

grid_search.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)

accuracy_score(y_pred_knn,y_test)
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(random_state=1)
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)

accuracy_score(y_pred_dt,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

dt=DecisionTreeClassifier(random_state=1)

dt_params={'max_depth':range(2,12),'max_leaf_nodes': list(range(2, 20)), 'min_samples_split': [2, 3]}

grid_search_dt=GridSearchCV(dt,dt_params,cv=10)

grid_search_dt.fit(X_train,y_train)

grid_search_dt.best_params_
```

```python
dt=DecisionTreeClassifier(random_state=1,max_depth=5,max_leaf_nodes=7,min_samples_split=2)
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)

accuracy_score(y_pred_dt,y_test)
```

## Naive bayes

```python
from sklearn.naive_bayes import GaussianNB

gn=GaussianNB()
gn.fit(X_train,y_train)
y_pred_gn=gn.predict(X_test)

accuracy_score(y_pred_gn,y_test)
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression(random_state=1,penalty='l2')
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)

accuracy_score(y_pred_lr,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

lr=LogisticRegression(random_state=1)

lr_params={ "penalty":["l1","l2"]}

grid_search_lr=GridSearchCV(lr,lr_params,cv=10)

grid_search_lr.fit(X_train,y_train)

grid_search_lr.best_params_
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression(random_state=1,penalty='l2')
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)

accuracy_score(y_pred_lr,y_test)
```

## support vector classifier

```python
from sklearn.svm import SVC

sv=SVC(kernel='sigmoid',random_state=1)
sv.fit(X_train,y_train)
y_pred_svc=sv.predict(X_test)

accuracy_score(y_pred_svc,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

sv=SVC()

svc_params = [{'kernel':['rbf','sigmoid'],'degree':list(range(2,4))}]
    

grid_search_svc = GridSearchCV(sv,svc_params, cv = 10, scoring='accuracy')

grid_search_svc .fit(X_train, y_train)

grid_search_svc.best_params_
```

```python
sv=SVC(kernel='rbf',degree=2)
sv.fit(X_train,y_train)
y_pred_svc=sv.predict(X_test)

accuracy_score(y_pred_svc,y_test)
```
